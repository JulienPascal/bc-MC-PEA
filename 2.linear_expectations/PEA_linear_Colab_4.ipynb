# I. Theory

## I.1. Model description

Let us consider a simple forward-looking expectational difference equation  
\begin{equation}
y_t = a\,\mathbb{E}_t[y_{t+1}] + b\,x_t,
\end{equation}

where the state $x_t$ follows an AR(1) with mean $\bar x$:  
\begin{equation}
x_{t+1} = \rho\,x_t + (1-\rho)\,\bar x + \varepsilon_{t+1},
\quad |\rho|<1.
\end{equation}

Parameters $a,b,\rho,\bar x$ are given, and $\varepsilon_t \sim \mathcal N(0,\sigma_{\varepsilon}^2)$.

---

## I.2. Analytic solution for $y_t$

1. **Forward-sum representation**  
   Under the no-bubble solution (transversality condition)  
   \begin{equation}
   y_t = b \sum_{i=0}^\infty a^i\,\mathbb{E}_t[x_{t+i}].
   \end{equation}

2. **Compute $\mathbb{E}_t[x_{t+i}]$**

   For the AR(1) with intercept,  
   \begin{equation}
   \mathbb{E}_t[x_{t+i}]
   = \rho^i\,x_t + (1-\rho^i)\,\bar x.
   \end{equation}

3. **Evaluate the two infinite sums**  
   - Slope term:  
     \begin{equation}
     \sum_{i=0}^\infty (a\rho)^i = \frac{1}{1 - a\,\rho}.
     \end{equation}
   - Intercept term:  
     \begin{equation}
       \sum_{i=0}^\infty a^i(1-\rho^i)
       = \frac{1}{1-a} - \frac{1}{1-a\rho}
       = \frac{a\,(1-\rho)}{(1-a)\,(1-a\rho)}.
     \end{equation}

4. **Closed-form**  
   Plugging in,  
   \begin{equation}
   y_t
   = b\,x_t\;\frac{1}{1 - a\,\rho}
     + b\,\bar x\;\frac{a\,(1-\rho)}{(1-a)\,(1-a\rho)},
   \end{equation}
   i.e.
   \begin{equation}
   \boxed{
     y_t
     = \frac{b}{1 - a\,\rho}\,x_t
     + \frac{a\,b\,(1-\rho)}{(1-a)\,(1-a\rho)}\,\bar x
   }.
   \end{equation}

---

## I.3. Analytic form of $\mathbb{E}_t[y_{t+1}]$

Let's make the parametric assumption for the conditional expectation term:
\begin{equation}
\mathbb{E}_t[y_{t+1}] \approx \theta_0 + \theta_1\,x_t.
\end{equation}

1. **Use the solution for $y_{t+1}$**  
   \begin{equation}
   y_{t+1}
   = \frac{b}{1 - a\,\rho}\,x_{t+1}
     + \frac{a\,b\,(1-\rho)}{(1-a)\,(1-a\rho)}\,\bar x.
   \end{equation}

2. **Take conditional expectation**  
   Since $\mathbb{E}_t[x_{t+1}] = \rho\,x_t + (1-\rho)\,\bar x$  
   \begin{equation}
   \mathbb{E}_t[y_{t+1}]
   = \frac{b}{1 - a\,\rho}\bigl[\rho\,x_t+(1-\rho)\bar x\bigr]
     + \frac{a\,b\,(1-\rho)}{(1-a)\,(1-a\rho)}\,\bar x.
   \end{equation}

3. **Collect terms**  
   \begin{equation}
   \theta_1 = \frac{b\,\rho}{1 - a\,\rho},
   \quad
   \theta_0 = \frac{b\,(1-\rho)}{(1-a)\,(1-a\rho)}\,\bar x.
   \end{equation}

Hence  
\begin{equation}
\boxed{
  \mathbb{E}_t[y_{t+1}]
  = \frac{b\,(1-\rho)}{(1-a)\,(1-a\rho)}\,\bar x
    + \frac{b\,\rho}{1 - a\,\rho}\,x_t
}.
\end{equation}

With the **PEA**, we hope to estimate the parameters using OLS:
\begin{equation}
\begin{split}
\hat{\theta_0} & = \frac{b\,(1-\rho)}{(1-a)\,(1-a\rho)}\,\bar x \\
\hat{\theta_1} & = \frac{b\,\rho}{1 - a\,\rho}
\end{split}
\end{equation}

---

## I.4. An economic interpretation

Some economic models fit within the simple linear rational expectations model from above.
One example is the **Cagan model of hyperinflation**.

Consider an economy in which money demand (= money supply in equilibrium), divided by the price level, depends negatively on expected inflation:

\begin{equation}
\frac{M_t}{P_t} = \exp(- \alpha \pi_{t+1}^e)
\end{equation}

where $\pi_{t+1}^e = E_t[\frac{P_{t+1} - P_{t}}{P_t}]$.

Let's use non-capitalized letters to denote the logarithm of terms (e.g., $p_t = \log(P_t))$.

Observe that  $\pi_{t+1}^e = E_t[\exp(p_{t+1}^e - p_t) - 1] \approx E_t[p_{t+1}^e - p_t]$, using $\exp(x) \approx 1 + x$ for $x$ small.

Hence, we have:

\begin{equation}
m_t - p_t \approx - \alpha  ( E_t[p_{t+1}^e ] - p_t)
\end{equation}

or, equivalently:

\begin{equation}
p_t \approx \frac{\alpha}{1 + \alpha } E_t[p_{t+1}^e ] + \frac{1}{1 + \alpha} m_t
\end{equation}

Define $a = \frac{\alpha}{1 + \alpha }$ and $ b = \frac{1}{1 + \alpha}$, then we have (forgetting the approximation):

\begin{equation}
p_t = a E_t[p_{t+1}^e ] + b m_t
\end{equation}

For the **law of motion of money supply**, we may assume that the (detrended, removing a constant growth term) log of money supply follows an AR(1) process:

\begin{equation}
m_{t+1} = \rho\,m_t + (1-\rho)\,\bar m + \varepsilon_{t+1},
\quad |\rho|<1.
\end{equation}

with $\varepsilon_t \sim \mathcal N(0,\sigma^2)$.

---

## I.5. Convergence Analysis

Let's analyze the convergence of the PEA, for $\bar{x} = 0$, to make our life simpler. With that extra assumption, the law of motion for the state vector simplifies to :

\begin{equation}
x_{t+1} = \rho\,x_t +  \varepsilon_{t+1},
\quad |\rho|<1.
\end{equation}

We start with an initial guess $\theta_0^{(0)} = \theta_0^{0}$ and $\theta_1^{(0)} = \theta_1^{0}$.

At step $k$ of the PEA, our guess for the expectation function is $E_t[y_{t+1}] = \theta_0^{(k)} + \theta_1^{(k)} x_t$. Because $y_t = a E_t[y_{t+1}] + b x_t$, we get:

\begin{equation}
y_t = a \theta_0^{(k)} + ( a \theta_1^{(k)} + b) x_t
\end{equation}


Forward $y_t$ by one period and use the law of motion for $x_{t+1}$ to get:

\begin{equation}
y_{t+1} = a \theta_0^{(k)} + \rho ( a \theta_1^{(k)} + b) x_t + \varepsilon_{t+1}
\end{equation}

OLS is unbiased, so at step $k+1$ and assuming full update ($\tau = 1$), we get:

\begin{equation}
\begin{split}
\theta_0^{(k+1)} & =  a \theta_0^{(k)} \\
\theta_1^{(k+1)} & = \rho ( a \theta_1^{(k)} + b)
\end{split}
\end{equation}

It is easy to see that, as long as $|a|<1$ and $|\rho a| < 1$, we have a contraction mapping, leading to the fixed point:

\begin{equation}
\begin{split}
\theta_0^{*} & =  0 \\
\theta_1^{*} & = \frac{b\,\rho}{1 - a\,\rho}
\end{split}
\end{equation}

Thus, the PEA converges to the true solution. We can even calculate the expected values for $\theta_0^{(k)}$ and $\theta_1^{(k)}$ after $k$ updates:

\begin{equation}
\begin{split}
\theta_0^{(k)} & =  a^{k} \theta_0^{0} \\
\theta_1^{(k)} & = \frac{b \rho}{1 - a \rho } \big(1 - (a \rho)^{n} \big) +  \theta_1^{0} (a \rho)^{n}
\end{split}
\end{equation}

### II.B. Standard PEA

PEA for $y_t = a E_t[y_{t+1}] + b x_t$

with $x_{t+1} = \rho x_t + (1 - \rho) \bar{x} + \varepsilon_{t+1}$

Approximates $E_t[y_{t+1}] ≈ \theta_0 + \theta_1 * x_t$


```python
# Analytic solutions:
def analytic_coefs_yt(a, b, rho, xbar):
    """
    Analytic no-bubble solution parameters for y_{t} = theta0_tilde + theta1_tilde * x_t.
    Valid for x_{t+1} = rho * x_t + (1 - rho) * xbar + eps_{t+1}.
    """
    theta1_tilde = b / (1 - a * rho)
    theta0_tilde = (a * b * (1 - rho) * xbar) / ((1 - a) * (1 - a * rho))
    return theta0_tilde, theta1_tilde

def analytic_theta(a, b, rho, xbar):
    """
    Analytic no-bubble solution parameters for E_t[y_{t+1}] = theta0 + theta1 * x_t.
    Valid for x_{t+1} = rho * x_t + (1 - rho) * xbar + eps_{t+1}.
    """
    theta1 = rho * b / (1 - a * rho)
    theta0 = xbar * b * (1 - rho) / ((1 - a)*(1 - a * rho))
    return theta0, theta1
```

```python
# Generate draws
def generate_random_innovations(slong, std, sampler = None, random_permutations = False):
    """
    Generate 1-D innovations ~ N(0, std^2) of length slong.
    If `sampler` is None, random draws.
    Otherwise draw using `sampler.random(n_samples)`, where sampler is expected to be a low-discrepancy sampler.
    """
    if sampler is None:
        e = std * np.random.randn(slong)
    else:
        # 1. Draws from QMC
        e = sampler.random(slong).reshape(-1)
        # 2. Break serial correlation by shuffling
        if random_permutations == True:
            perm = np.random.default_rng().permutation(slong)
            e = e[perm]
    return e

def generate_random_arrays(slong, N, std, sampler = None, random_permutations = False):
    """
    Generate:
      • e: a 1-D array of length `slong` with entries ~ N(0, std^2)
      • E: a (slong × N) random matrix. Each ROW is a quasi Monte Carlo sample ~ N(0, std^2)
    """
    # vectorized draws in one go:
    if sampler is None:
        e = std * np.random.randn(slong)
        E = std * np.random.randn(slong, N)
    else:
        e = std * np.random.randn(slong) # Generate random draws, to push forward the simulation
        E = generate_random_innovations(slong * N, std, sampler, random_permutations).reshape(slong, N)

    return e, E

def generate_ld_sampler(params):
    """
    Generate a low-discrepancy sampler if params.low_discrepancy is True.
    Otherwise, returns None

    """
    # If low-discrepancy, build a sampler for innovation vectors
    if params.low_discrepancy == True:
        #sob = qmc.Sobol(d=1, scramble=True) ## Not needed
        ld_sampler = qmc.MultivariateNormalQMC(mean=[0], cov=[[params.std_e**2]])
    else:
        ld_sampler = None

    return ld_sampler


def gamma_k(k, K=50, c=0.5):
    """
    Compute step size γ_k for PEA update:
      γ_k = 1       if k ≤ K
      γ_k = K / k   if k > K

    Parameters:
    - k: iteration number (1-based integer)
    - K: threshold for full-weight updates (default 50)

    Returns:
    - γ_k: float
    """
    #return 1.0 if k <= K else K / k
    #return 1/(k**c)
    return 1.0 if k <= K else K/(k**c)
```

```python
# PEA params
T=10000 #number of draws for the state vector. One innovation draw per shock.
burn=500 #burin period
tol=1e-6 #tolerance on two guesses for the parameter vector
max_iter=1000 #maximum number of iterations
gamma = 1.0 #1.0 #convex parameter for the update of guesses
new_shock_every= 1000 #draws new shocks every x iterations
time_varying_gamma = False #True # schedule for updates
seed=123 #seed number

# Initial guess for [theta0, theta1]
theta_init = np.array([0.25, 0.5]) #np.array([0.5, 1.0])

theta = theta_init.copy()
theta_hist = [theta.copy()] #store history for parameter vector
diff_hist = [] #store history of |theta^{n+1} - theta^{n}|

eps = generate_random_innovations(T+1, params.std_e) #rng.standard_normal(T + 1) * sigma_e #innovation draws
x = np.empty(T+1) #state vector

np.random.seed(seed)
check_val = 1000.0

for it in range(1, max_iter + 1):
    # redraw shocks
    if it % new_shock_every == 0:
         eps[:] = generate_random_innovations(T+1, params.std_e)

    # Simulate AR(1) x-path
    x[0] = params.mean_x
    for t in range(T):
        x[t + 1] = params.rho_x * x[t] + (1 - params.rho_x) * params.x_bar + eps[t + 1]

    # Simulate y-path using current theta
    y = np.zeros(T + 1)
    for t in range(T+1):
        Et_y = theta[0] + theta[1] * x[t]
        y[t] = params.a * Et_y + params.b * x[t]

    # OLS regression to update theta: E_{yt+1} = x'theta
    X = np.column_stack([np.ones(T), x[:-1]]) #remove last period
    Y = y[1:] #forward by one
    Xr, Yr = X[burn:], Y[burn:]
    theta_new, _, _, _ = np.linalg.lstsq(Xr, Yr, rcond=None)

    theta_hist.append(theta_new.copy())

    # Convergence check
    check_val = np.max(np.abs(theta_new - theta))
    diff_hist.append(check_val)
    if check_val < tol:
        print(f"PEA converged in {it} iterations. Tolerance: {check_val}")
        theta = theta_new
        break

    # Damped update
    theta = gamma * theta_new + (1 - gamma) * theta

    # Update gamma parameter:
    if time_varying_gamma == True:
      gamma = gamma_k(it)

else:
    print(f"Warning: PEA did not converge within max_iter: {max_iter}. Tolerance {check_val}. Update size: {gamma}")

# residual
Res = Yr - np.dot(Xr, theta)
MSE = np.mean(Res ** 2)

# 1. Plot histogram of residuals
plt.figure()
plt.hist(Res, bins=100)
plt.title('Residuals', fontname='Times', fontsize=12)
plt.show()

# Store values
theta_history = np.array(theta_hist)
theta_est = theta.copy()

# Compute analytic solution
theta0_true, theta1_true = analytic_theta(params.a, params.b, params.rho_x, params.x_bar)

# Print comparison
print(f"Estimated theta0 = {theta_est[0]:.6f}, analytic = {theta0_true:.6f}, error = {theta_est[0] - theta0_true:.2e}")
print(f"Estimated theta1 = {theta_est[1]:.6f}, analytic = {theta1_true:.6f}, error = {theta_est[1] - theta1_true:.2e}")

# Plot convergence of theta estimates
iterations = np.arange(len(theta_history))
plt.figure(figsize=(8, 4))
plt.plot(iterations, theta_history[:, 0], label=r'$\theta_0$ est')
plt.axhline(theta0_true, color='k', linestyle='--', label=r'$\theta_0$ true')
plt.plot(iterations, theta_history[:, 1], label=r'$\theta_1$ est')
plt.axhline(theta1_true, color='gray', linestyle='--', label=r'$\theta_1$ true')
plt.xlabel('Iteration')
plt.ylabel('Theta values')
plt.title('PEA Convergence: Estimated vs. True Theta')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 4))
plt.plot(diff_hist)
plt.xlabel('Iteration')
plt.ylabel('Difference thetas')
plt.yscale('log')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## II.C. bc-MC-PEA

Use multiple innovation draws per realization of the state vector.

```python
@njit
def simulate_path_N_inplace(
    M: int, #number of draws in the state space
    N: int, #number of innovation draws, per state vector draw
    init: int, #number of draws discarded
    mean_x: float, #value value for x
    std_x: float, # std dev. for x
    rho_x: float, #persistence parameter for x
    x_bar: float, #mean for x
    a: float,
    b: float,
    e: np.ndarray, #i.i.d. draws for innovation
    E: np.ndarray, #extra draws for innovation
    b0: np.ndarray, #parameter vector
    nb_expl_vars: int,
    basis: int,
    x: np.ndarray,
    y: np.ndarray,
    X: np.ndarray,
    y_temp: np.ndarray
) -> None:
    """
    Simulates one path using a Parameterized Expectations Algorithm (PEA)
    with in-place mutation of pre-allocated arrays. JIT-compiled with numba for maximum performance.

    All outputs are written directly into the provided arrays; no value is returned.
    """
    slong = M + init

    # Set value for constant vector
    X[:, 0] = 1.0

    # Unpack coefficients for inlined dot-products
    b0_0 = b0[0]; b0_1 = b0[1]

    # Initialize state
    x[0] = mean_x
    # Simulate AR(1) x-path
    for t in range(1, slong):
        x[t] = rho_x * x[t-1] + (1 - rho_x) * x_bar + e[t]

    for i in range(0, slong):
        # Fill in X[i] for OLS regression y = X*b
        X[i, 1] = x[i]

        # Only do multi-innovation MC for i >= init. No need for burnin phase.
        if i >= init:
            # Monte Carlo expectation
            for j in range(N):
                # forward the state by one period, using different innovations
                x_next = rho_x * x[i] + (1 - rho_x) * x_bar + E[i, j]

                # Linear model for conditional expectation
                ## E_t[y_{t+1}], for several realization of the innovation vector
                #
                # E_t[y_{t+1}] = b0_0 + b0_1 * x_t, and y_t = a E_t[y_{t+1}] + b * x_t
                # gives: y_t = a (b0_0 + b0_1 * x_t) + b * x_t
                y_temp[j] = a * b0_0 + (a*b0_1 + b)* x_next

            # Final aggregation, to calculate Et_{yt+1} as an average
            s = 0.0
            for j in range(N):
                s += y_temp[j]

            y[i] = (s / N)

#@njit
def calculate_MSIE(
    slong: int, #number of draws in the state space
    init: int, #number of draws discarded
    mean_x: float, #value value for x
    std_x: float, # std dev. for x
    rho_x: float, #persistence parameter for x
    x_bar: float, #mean for x
    e: np.ndarray, #i.i.d. draws for innovation
    b0_true: np.ndarray, #True parameter vector
    b0: np.ndarray, #Estimate parameter vector
):
    """
    Calculate mean squared integration error using Monte Carlo integration.
    """
    MSIE_array = np.zeros(slong)
    x = np.zeros(slong+1)

    # Unpack coefficients for inlined dot-products
    b0_0_diff = b0_true[0] - b0[0]
    b0_1_diff = b0_true[1] - b0[1]

    # Initialize state
    x[0] = mean_x
    # Simulate AR(1) x-path
    for t in range(1, slong):
        x[t] = rho_x * x[t-1] + (1 - rho_x) * x_bar + e[t]

    # Calculate squared integration error:
    for t in range(0, slong):
        MSIE_array[t] = (b0_0_diff + b0_1_diff * x[t])**2

    MSIE = np.mean(MSIE_array[init:-1]) #remove last observation
    return MSIE

def calculate_MSIE_analytic(
    mean_x: float, #value value for x
    std_x: float, # std dev. for x
    rho_x: float, #persistence parameter for x
    x_bar: float, #mean for x
    b0_true: np.ndarray, #True parameter vector
    b0: np.ndarray, #Estimate parameter vector
):
    """
    Calculate mean squared integration error using analytic formula
    """
    # Unpack coefficients for inlined dot-products
    Δ0  = b0_true[0] - b0[0]
    Δ1  = b0_true[1] - b0[1]

    # Obtained by expanding the square and using the fact that x_t is AR(1)
    MSIE = (Δ0**2 + 2*Δ0*Δ1*x_bar + Δ1**2*(x_bar**2 + std_x**2))

    return MSIE

```

```python
%%time

# --------------------------------------------------------------------------
# Main code (outside numba) to run the PEA iterations.
# --------------------------------------------------------------------------
list_N = [1, 2, 5, 10, 25, 50, 100, 250, 500]
MSE_N = np.zeros(len(list_N)) #In-sample
MSIE_N = np.zeros(len(list_N)) #Mean square integration error, out-sample
MSIE_N_analytic = np.zeros(len(list_N)) #Mean square integration error, out-sample
Time_N = np.zeros(len(list_N))

M = 10000
init = 100
slong = init + M

# Compute analytic solution
theta0_true, theta1_true = analytic_theta(params.a, params.b, params.rho_x, params.x_bar)
b0_true = np.array([theta0_true, theta1_true])

# To store errors on parameter vector
list_err_theta0 = []
list_err_theta1 = []

tol = 1e-8
gam = 1.0  # smoothing parameter
max_iter = 100 # max number of iterations
redraw_shocks_every = 1000 #redraw new realizations of innovations (new state and innovation vectors)

slong_test = 100000
init_test = 100
e_test =  generate_random_innovations(slong_test, params.std_e)

np.random.seed(1)
#for N in range(1, max_N+1):
for (index_N, N) in enumerate(list_N):
    print("Running simulation with N =", N)

    b0_current = theta_init.copy()  # initial guess (shape (6,))

    # Generate a ld sampler. If random draws, ld_sampler is None
    ld_sampler = generate_ld_sampler(params)

    # innovation for state vector
    e, E = generate_random_arrays(slong, N, params.std_e, ld_sampler)

    # Preallocate arrays
    x = np.zeros(slong+1)
    y_out      = np.zeros(slong)        # simulated y
    X_data      = np.zeros((slong, params.nb_expl_vars))   # regressor matrix
    y_temp = np.zeros(N)            # temporary array for innovations

    # Warmup (compilation) first go
    simulate_path_N_inplace(M, N, init, params.mean_x, params.std_x, params.rho_x, params.x_bar,
                        params.a, params.b, e, E, b0_current,
                        params.nb_expl_vars, params.basis,
                        x, y_out, X_data,y_temp)

    start_time = time.perf_counter()
    iter_num = 1
    crit = 1.0
    # Run a fixed number of iterations (or use while crit > tol)
    while iter_num < max_iter:
        if iter_num % redraw_shocks_every == 0:
            e, E = generate_random_arrays(slong, N, params.std_e, ld_sampler)

        # Call the numba-compiled simulation to get regression data.
        simulate_path_N_inplace(M, N, init, params.mean_x, params.std_x, params.rho_x, params.x_bar,
                        params.a, params.b, e, E, b0_current,
                        params.nb_expl_vars, params.basis,
                        x, y_out, X_data, y_temp)

        # Remove burnin
        X_reg = X_data[init:, :]
        y_reg = y_out[init:]

        # OLS
        bt, _, _, _ = np.linalg.lstsq(X_reg, y_reg, rcond=None)

        # Parameter update
        b_new = gam * bt + (1 - gam) * b0_current
        crit = np.max(np.abs(b_new - b0_current))
        b0_current = b_new.copy()
        if iter_num % 10 == 0:
            print("Iteration:", iter_num, "Conv. crit.:", crit)
        iter_num += 1

    end_time = time.perf_counter()
    elapsed = end_time - start_time
    print("Elapsed time for N =", N, ":", elapsed, "seconds")

    # Compute residuals and In-sample MSE.
    Res = y_reg - np.dot(X_reg, b0_current)
    MSE = np.mean(Res ** 2)
    MSE_N[index_N] = MSE

    # Error on parameter values:
    err_theta0 = b0_current[0] - theta0_true
    err_theta1 = b0_current[1] - theta1_true
    list_err_theta0.append(err_theta0)
    list_err_theta1.append(err_theta1)

    MSIE_N[index_N] = calculate_MSIE(slong_test, init_test, params.mean_x,
               params.std_x, params.rho_x, params.x_bar,
               e_test, b0_true, b0_current)

    MSIE_N_analytic[index_N] = calculate_MSIE_analytic(params.mean_x,
               params.std_x, params.rho_x, params.x_bar,
               b0_true, b0_current)

    Time_N[index_N] = elapsed
    print("Final iteration N:", N, "Iterations:", iter_num, "OLS MSE:", MSE)
    print("Final b0:", b0_current)
```

```python
def compute_pareto_front(df, time_col='Time', mse_col='MSIE'):
    # Sort by Time ascending
    df_sorted = df.sort_values(by=time_col, ascending=True)

    # List to store the Pareto front
    pareto_points = []

    # Keep track of the lowest MSE encountered so far
    best_mse_so_far = float('inf')

    for idx, row in df_sorted.iterrows():
        mse_val = row[mse_col]
        if mse_val < best_mse_so_far:
            pareto_points.append(row)
            best_mse_so_far = mse_val

    return pd.DataFrame(pareto_points)

pareto_df = compute_pareto_front(df_results_average)

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_results_average, x='Time', y='MSIE', hue='M', palette='viridis', alpha=0.7)
plt.plot(pareto_df['Time'], pareto_df['MSIE'], 'r-o', label='Pareto Frontier')

# Now highlight the optimal points
optimal_df = df_results_average[df_results_average['N'] == 1]
plt.scatter(optimal_df['Time'], optimal_df['MSIE'],
            marker='+', s=250, c='red',
            edgecolors='black', linewidths=1,
            label='N = 1')

plt.xlabel('Time (seconds)')
plt.ylabel('MSIE')
plt.yscale('log')
plt.xscale('log')
plt.title('Time vs. MSIE, Colored by M')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(8,6))
sns.scatterplot(data=df_results_average, x='Time', y='MSIE', hue='N', palette='viridis', alpha=0.7)
plt.plot(pareto_df['Time'], pareto_df['MSIE'], 'r-o', label='Pareto Frontier')

# Now highlight the optimal points
optimal_df = df_results_average[df_results_average['N'] == 1]
plt.scatter(optimal_df['Time'], optimal_df['MSIE'],
            marker='+', s=250, c='red',
            edgecolors='black', linewidths=1,
            label='N = 1')

plt.xlabel('Time (seconds)')
plt.ylabel('MSIE')
plt.yscale('log')
plt.xscale('log')
plt.title('Time vs. MSIE, Colored by N')
plt.legend()
plt.grid(True)
plt.show()
```

#### Uncertainty quantification

```python

def optimal_M_N(T, k, alpha, alpha0, alpha_MN, alpha_M, eps=1e-12, round_nearest = 1):
  """
  Compute the pair (M*, N*) that minimizes MSIE
    E[(…)^2] ∼ σ^2 k / [N^α (M−k−1)]
  subject to the time budget
    T = α0 + α_{MN}⋅M⋅N + α_M⋅M.

  Parameters
  ----------
  T : float
      Total time budget (must exceed α0).
  k : int
      Number of regressors, including constant
  alpha : float
      Exponent in the integration error ( >1 for well‑posed optimum).
  alpha0 : float
      Intercept term (c_st).
  alpha_MN : float
      Coefficient on M⋅N.
  alpha_M : float
      Coefficient on M.
  eps : float, optional
      Threshold below which we treat alpha2 as zero.
  round_nearest : int, optional
      rount to nearest multiple of round_nearest. By default, one

  Returns
  -------
  M_star, N_star : tuple of floats
      Optimal M and N.
  """
  C0 = T - alpha0
  A = (alpha - 1) * C0
  B = alpha * C0 * (k + 1)

  if C0 <= 0:
      raise ValueError("T must exceed alpha0")
  # α2 → 0 limit
  if abs(alpha_M) < eps:
      #print(f"corner case, alpha_M: {alpha_M}")
      N_star = 1
      M_star = C0 / (alpha_MN * N_star)
  else:
      # positive root of α2 M^2 + (α-1)C0 M - α C0 (k+1) = 0
      M_star = (-A + np.sqrt(A**2 + 4 * alpha_M * B)) / (2 * alpha_M)
      # back out N from the time constraint
      N_star = (C0 - alpha_M * M_star) / (alpha_MN * M_star)

  # round to nearest value
  M_star = int(np.round(M_star/round_nearest) * round_nearest)
  N_star = int(np.round(N_star/round_nearest) * round_nearest)

  return M_star, N_star

# Take also into account uncertainty regarding the intercept
def se_MN_star_full(
    M_star, N_star, T, k, alpha,
    alpha_0, alpha_M, alpha_MN,
    se_alpha_0, se_alpha_M, se_alpha_MN,
    cov_mat          # 3×3 numpy array from m.cov_params().values
):
    """
    cov_mat should match the order of grad_M = np.array([dM_dalpha0, dM_dalphaM, dM_dalphaMN])
    """
    # Precompute
    C0 = T - alpha_0
    b  = (alpha - 1)*C0
    D  = np.sqrt(b*b + 4*alpha_M*alpha*C0*(k+1))

    # 1) partials for M*
    dM_d0 = -1/(2*alpha_M)*((alpha-1)
             - ((alpha-1)**2 *C0 + 2*alpha*alpha_M*(k+1))/D)
    dM_dM = (alpha*C0*(k+1)/D - M_star)/alpha_M
    dM_dMN= 0.0

    grad_M = np.array([dM_d0, dM_dM, dM_dMN])
    var_M  = grad_M @ cov_mat @ grad_M
    se_M   = np.sqrt(var_M)

    # 2) partials for N*
    dN_d0 = (1/(alpha_MN*M_star)
             - (C0 - alpha_M*M_star)/(alpha_MN*(M_star**2))*dM_d0)
    dN_dM = (-1/alpha_MN
             - (C0 - alpha_M*M_star)/(alpha_MN*(M_star**2))*dM_dM)
    dN_dMN= -N_star/alpha_MN

    grad_N = np.array([dN_d0, dN_dM, dN_dMN])
    var_N  = grad_N @ cov_mat @ grad_N
    se_N   = np.sqrt(var_N)

    return se_M, se_N

target_time = 1
for alpha_selected in [coef_alpha, 1.0]:
    M_opt, N_opt = optimal_M_N(target_time, params.nb_expl_vars, alpha_selected, c_st, alpha_MN, alpha_M)

    se_M_opt, se_N_opt = se_MN_star_full(M_opt, N_opt, target_time, params.nb_expl_vars, alpha_selected,
                                        c_st, alpha_M, alpha_MN,
                                        se_c_st, se_alpha_M, se_alpha_MN, cov_mat)
    print(f"alpha: {alpha_selected}")
    print(f"M*: {M_opt} +/- 2 * {se_M_opt}. || 2 * std dev.(M*) / M* : {2 * se_M_opt/M_opt}")
    print(f"N*: {N_opt} +/- 2 * {se_N_opt}. || 2 * std dev.(N*) / N* : {2 * se_N_opt/N_opt}")

```

```python
# Redo regression, but with data close to target times
threshold_reg = 10.0 #in seconds
min_M = 5 #Minimumm value for M. Otherwise blows up

df_selected = df_results_average[df_results_average["Time"] < threshold_reg]

# Accuracy regression
m = smf.ols("np.log(MSIE) ~ np.log(N) + np.log(M - k - 1)", data=df_selected).fit()

use_theoretical_value_alpha = True
if use_theoretical_value_alpha == True:
    if params.low_discrepancy == True:
        coef_alpha = 2.0 #Use theoretical value
    else:
        coef_alpha = 1.0
else:
    coef_alpha = -m.params["np.log(N)"] #Use estimated value

print(coef_alpha)
print(m.summary())

# Time regression
# O(MN) + O(M)
# 1. OLS
m = smf.ols("Time ~ I(M*N) + M ", data=df_selected).fit()
print(m.summary())

# 2. Extract point estimates
c_st = m.params["Intercept"]
alpha_MN = m.params["I(M * N)"]      # coeff on M*N
alpha_M  = m.params["M"]             # coeff on M

# 3. Extract standard errors
se_c_st = m.bse["Intercept"]
se_alpha_MN = m.bse["I(M * N)"]      # SE of alpha_MN
se_alpha_M  = m.bse["M"]             # SE of alpha_M

# 4. Extract covariance between the two slopes
cov_M_MN = m.cov_params().loc["M", "I(M * N)"]  # covariance between alpha_M and alpha_MN

# 5. Extra full covariance matrix
print(m.cov_params()) # full var-cov matrix
# Reorder
cov_mat = m.cov_params().loc[['Intercept','M','I(M * N)'], ['Intercept','M','I(M * N)']].values

print("alpha_MN =", alpha_MN, "±", se_alpha_MN)
print("alpha_M  =", alpha_M,  "±", se_alpha_M)

print("Cov-Mat", cov_mat)


list_N        = []
list_se_N     = []
list_M        = []
list_se_M     = []
list_optimal  = []
list_Time     = []

list_target_times = [0.25, 0.5, 1.0, 1.5, 2.0]

for target_time in list_target_times:
    # 1) build candidate grid
    list_N_temp = [1, 10, 100, 1000, 10000]
    N_array     = np.array(list_N_temp)

    # recover M from your linear timing law (without burn-in)
    vals = (target_time - c_st) / (alpha_M + alpha_MN * N_array)

    # 2) correct rounding: apply *10 before comprehension
    M_vals = np.round(vals/10) * 10
    list_M_temp = [int(max(m, min_M)) for m in M_vals]  # ensure M ≥ min_M

    # 3) append the non‐optimal candidates
    list_N       += list_N_temp
    list_M       += list_M_temp
    list_optimal += [0] * len(list_N_temp)
    list_Time    += [target_time] * len(list_N_temp)  # align one‐to‐one
    list_se_N += [0] * len(list_N_temp)
    list_se_M += [0] * len(list_M_temp)

    # 4) compute the theoretical optimum with burn-in
    M_opt, N_opt = optimal_M_N(target_time, params.nb_expl_vars, coef_alpha, c_st, alpha_MN, alpha_M, round_nearest=True)
    se_M_opt, se_N_opt = se_MN_star_full(M_opt, N_opt, target_time, params.nb_expl_vars, coef_alpha,
                                        c_st, alpha_M, alpha_MN,
                                        se_c_st, se_alpha_M, se_alpha_MN, cov_mat)

    ci_lower_M = M_opt - 2*se_M_opt
    ci_higher_M = M_opt + 2*se_M_opt

    ci_lower_N = N_opt - 2*se_N_opt
    ci_higher_N = N_opt + 2*se_N_opt

    print(f"alpha: {coef_alpha}")
    print(f"M*: {M_opt}. stdev. {se_M_opt}; ({ci_lower_M} - {ci_higher_M})")
    print(f"N*: {N_opt}. stdev. {se_N_opt}; ({ci_lower_N} - {ci_higher_N})")

    # 5) append the optimal point
    list_N.append(int(N_opt))
    list_M.append(int(M_opt))
    list_optimal.append(1)
    list_Time.append(target_time)
    list_se_N.append(se_N_opt)
    list_se_M.append(se_M_opt)

# 6) build DataFrame
df_MN = pd.DataFrame({
    "M":           list_M,
    "N":           list_N,
    "is_optimal":  list_optimal,
    "target_time": list_Time,
    "se_M":        list_se_M,
    "se_N":        list_se_N
})


#----------------------------------
tol = 1e-8 # tolerance on parameter vector
gam = 1.0  # smoothing parameter between two iterations
max_iter = 100 # max number of iterations
redraw_shocks_every = 1000 #redraw new realizations of innovations (new state and innovation vectors)

slong_test = 1000000
init_test = 100
nb_tot_reps = 100 #100 #nb of repetitions, to smooth out randomness and potential issues with measuring timing

# to store restults
results = []
np.random.seed(42)

for (N, M, is_optimal) in zip(list_N, list_M, list_optimal):
    for nb_rep in range(nb_tot_reps):
        #
        #np.random.seed(nb_rep)
        # Innovations for the out-sample test
        e_test = generate_random_innovations(slong_test, params.std_e)

        slong = init + M

        b0_current = theta_init.copy()

        # Generate a ld sampler. If random draws, ld_sampler is None
        ld_sampler = generate_ld_sampler(params)

        # innovation for state vector
        e, E = generate_random_arrays(slong, N, params.std_e, ld_sampler)

        # Preallocate arrays
        x = np.zeros(slong+1)
        y_out      = np.zeros(slong)        # simulated y
        X_data      = np.zeros((slong, params.nb_expl_vars))   # regressor matrix
        y_temp = np.zeros(N)            # temporary array for innovations

        # Warmup (compilation) first go
        simulate_path_N_inplace(M, N, init, params.mean_x, params.std_x, params.rho_x, params.x_bar,
                            params.a, params.b, e, E, b0_current,
                            params.nb_expl_vars, params.basis,
                            x, y_out, X_data, y_temp)

        start_time = time.perf_counter()
        iter_num = 1
        crit = 1.0
        # Run a fixed number of iterations (or use while crit > tol)
        while iter_num < max_iter:
            if iter_num % redraw_shocks_every == 0:
                e, E = generate_random_arrays(slong, N, params.std_e, ld_sampler)

            # Call the numba-compiled simulation to get regression data.
            simulate_path_N_inplace(M, N, init, params.mean_x, params.std_x, params.rho_x, params.x_bar,
                            params.a, params.b, e, E, b0_current,
                            params.nb_expl_vars, params.basis,
                            x, y_out, X_data, y_temp)

            # Remove burnin
            X_reg = X_data[init:, :]
            y_reg = y_out[init:]

            # OLS
            bt, _, _, _ = np.linalg.lstsq(X_reg, y_reg, rcond=None)

            # Parameter update
            b_new = gam * bt + (1 - gam) * b0_current
            crit = np.max(np.abs(b_new - b0_current))
            b0_current = b_new.copy()

            #print("Iteration:", iter_num, "Conv. crit.:", crit)
            iter_num += 1

        end_time = time.perf_counter()
        elapsed = end_time - start_time
        #print("Elapsed time for N =", N, ":", elapsed, "seconds")

        # Compute residuals and In-sample MSE.
        Res = y_reg - np.dot(X_reg, b0_current)
        MSE = np.mean(Res ** 2)

        # Error on parameter values:
        err_theta0 = np.abs(b0_current[0] - theta0_true)
        err_theta1 = np.abs(b0_current[1] - theta1_true)

        #MSIE = calculate_MSIE(slong_test, init_test, params.mean_x,
        #          params.std_x, params.rho_x, params.x_bar,
        #           e_test, b0_true, b0_current)

        MSIE = calculate_MSIE_analytic(params.mean_x,
               params.std_x, params.rho_x, params.x_bar,
               b0_true, b0_current)

        if nb_rep == (nb_tot_reps-1):
            print(f"Final iteration (M, N): ({M}, {N}). Iterations: {iter_num}. Time (s): {elapsed}. OLS MSE: {MSE} , MISE: {MSIE}")
            print("Final b0:", b0_current)

        # Store the results in a dictionary
        results.append({
            "repetition": nb_rep,
            "k": params.nb_expl_vars,
            "M": M,
            "N": N,
            "Time": elapsed,
            "MSE": MSE,
            "MSIE": MSIE,
            "err_theta0": err_theta0,
            "err_theta1": err_theta1,
            "err_theta": 0.5*err_theta0 + 0.5*err_theta1,
            "is_optimal": is_optimal
            })

# Create a Pandas DataFrame from the results
df_results_2 = pd.DataFrame(results)
print(df_results_2.head())
```
